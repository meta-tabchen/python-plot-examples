```python
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_plot = pd.DataFrame({"a": np.random.random(20), 
                        "b": np.random.random(20),
                        "c": np.random.random(20)})
```

```python
col_map = {
    "a": r"$\sigma{(\alpha)}$",
    "b": r"$\sigma{(\beta)}$",
    "c": r"$\sigma{(\zeta)}$",
}
color_map = dict(zip(['a','b','c'],sns.color_palette(n_colors=3)))
#颜色参考: https://seaborn.pydata.org/generated/seaborn.color_palette.html
```

```python
sns.set_theme(context='notebook',style='white',font_scale=1.5,font='Times New Roman')
```

```python
fig = plt.figure(figsize=(12, 3))
x = list(range(len(df_plot)))

marker_list = ['o', 'o', 'D']
# markder参考: https://matplotlib.org/stable/api/markers_api.html
linestyle_list = ['dotted', 'dashed', 'solid']
# 线的类型参考,https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
for i, col in enumerate(['a', 'b', 'c']):
    ax = sns.lineplot(x=x, y=df_plot[col],
                      label=col_map.get(col, col),
                      marker=marker_list[i],
                      markersize=12,
                      linewidth=3,
                      linestyle=linestyle_list[i],
                      color=color_map[col],
                      legend='auto')
    sns.despine(left=False, bottom=False)  # 移除边框线
plt.ylabel("Score")
plt.xlabel("Length of student's interaction sequence")
# plt.title("Length")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.3), ncol=3)
```
