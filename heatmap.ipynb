## 画图示例

[API文档](https://seaborn.pydata.org/generated/seaborn.heatmap.html)

**sns.heatmap**
- **xticklabels**: x轴tick标尺显示的数字
- **yticklabels**: y轴tick显示的数字
- **annot**: 是否需要显示具体数字
- **fmt**: String formatting code to use when adding annotations. Such as '.4f'
- **cbar_kws**: 用来缩放bar{"shrink": 0.8},使其可以和
- **cmap**: The mapping from data values to color space.
- **square**: 是否为正方形

**plt.savefig**

[API文档](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html)

- **bbox_inches**: bbox_inches='tight' represents no bounding box around output image.
- **dpi**: 清晰度

[sns.set API](https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme)

```python
sns.set(context='notebook',font_scale=1.2)#设置字体缩放
```

```python
xticklabels = glue.columns.tolist()
print(f"xticklabels is {xticklabels}")
yticklabels = glue.index.tolist()
print(f"yticklabels is {yticklabels}")
```

```python
fig = plt.figure(figsize=(8, 8))
sns.heatmap(glue, square=True, xticklabels=xticklabels,
            yticklabels=yticklabels, fmt='.2f', annot=True,
            cbar_kws={"shrink": 0.82}, cmap="Blues_r")
plt.xlabel("Number of Up Sentences")  # 设置 X 轴的label
plt.ylabel("Number of Down Sentences")  # 设置Y轴的label
plt.title("Title")  # 设置标题
# plt.close(fig)
plt.savefig(f"imgs/heatmap/demo.jpg",bbox_inches="tight", dpi=100)
```
